## Check out annual flow metrics

```python
metric_change0 = pd.read_csv('../Data/metric_change_by_scenario_wet030722.csv')
metric_change1 = pd.read_csv('../Data/metric_change_by_scenario_avg030722.csv')
metric_change2 = pd.read_csv('../Data/metric_change_by_scenario_dry030722.csv')
metric_changeQ = pd.concat([metric_change0[metric_change0.type=='Q'],
                           metric_change1[metric_change1.type=='Q'],
                           metric_change2[metric_change2.type=='Q']])

metric_change = metric_changeQ
```

```python
# # number of indicators that change by by year 50 at each site
change_thresh = 10 # % difference between year 1 and 50

mean_change = metric_change.groupby(['site','climate','pumprate','metric']).mean().reset_index()
mean_change['change50'] = np.where(mean_change.perc_change_yr_50.abs()>change_thresh,1,0)
mean_change = mean_change.groupby(['site','climate','pumprate']).sum().drop(columns='perc_change_yr_50').reset_index()

fig,axs = plt.subplots(1,3,figsize = (18,6),sharex=True,sharey=True)

axes = [axs[0],axs[1],axs[2],
        axs[0],axs[1],axs[2]]

bins = np.linspace(0,15,8)
i=0
for p in mean_change.pumprate.drop_duplicates():
    for clim in mean_change.climate.drop_duplicates():
        mean_change[(mean_change.climate==clim) & 
                    (mean_change.pumprate==p)].change50.hist(ax=axes[i],bins = bins,edgecolor = 'k',
                                                            label = p,alpha=0.5)
        axes[i].set_title('Climate = '+clim+', pump schedule = '+p)
        axes[i].legend()
        if i>2:
            axes[i].set_xlabel('Number of metrics changed\nbetween year 1 and 50')
        i+=1
axs[0].set_ylabel('Number of sites')
# axs[1].set_ylabel('Number of sites')

fig.suptitle('Annual flow metrics')
```

```python
fig,ax = plt.subplots(1,figsize = (8,8))
bins = np.linspace(0,15,12)
i=0
p = 'Constant'
for clim in ['Dry','Avg','Wet']:
    ax.boxplot(mean_change[(mean_change.climate==clim) & 
                (mean_change.pumprate==p)].change50,
              positions=[i],
              boxprops= dict(linewidth=2.0, color='black'),
           whiskerprops = dict(linewidth=2.0, color='black'),
           medianprops = dict(linewidth=2.0, color='firebrick'),
           capprops = dict(linewidth=2.0, color='black'))
    i+=1

ax.set_ylabel('Number of metrics changed at each site\nbetween year 1 and 50')
ax.set_xlabel('Climate scenario')
ax.set_xticklabels(['Dry','Avg','Wet'])
ax.set_title('Annual flow metrics')
```

```python
gagesII = gpd.read_file('../Data/gagesII_data_and_subsets/gagesII.shp')
gagesII['site'] = gagesII.STAID.astype('int')
state_outline = gpd.read_file('../Data/maps_and_boundaries/US_state_outline/cb_2018_us_state_500k.shp').to_crs(epsg=4326)
```

```python
mean_change_sub = mean_change[(mean_change.climate=='Dry')&(mean_change.pumprate=='Constant')]
mean_change_sub = mean_change_sub.merge(gagesII[['site','LAT_GAGE','LNG_GAGE']],left_on='site',right_on='site')

fig,ax = plt.subplots(1,figsize = (15,8))
state_outline.boundary.plot(ax=ax,edgecolor = 'k',zorder = -1)
im = ax.scatter(mean_change_sub.LNG_GAGE,
           mean_change_sub.LAT_GAGE,
           c=mean_change_sub.change50,
           edgecolor = 'k',
           s = 150)
ax.set_xlim(-130,-65)
ax.set_ylim(22,51)
ax.set_title('Number of annual flow metrics changed')
plt.colorbar(im)
```

```python
# # number of sites for which each indicator changes by 10% by year 50
# # number of indicators that change by by year 50 at each site
change_thresh = 10 # % difference between year 1 and 50

mean_change = metric_change.groupby(['site','climate','pumprate','metric']).mean().reset_index()
mean_change['change50'] = np.where(mean_change.perc_change_yr_50>change_thresh,1,0)
mean_change = mean_change.groupby(['metric','climate','pumprate']).sum().drop(columns=['perc_change_yr_50','site']).reset_index()

metric_list = mean_change.metric.drop_duplicates()

fig,axs = plt.subplots(1,figsize = (15,6))
bins = np.linspace(0,15,12)
i=0
p = 'Constant'
for clim in ['Dry']:
    j=0
    subset =  metric_change[(metric_change.climate==clim) & 
                    (metric_change.pumprate==p) ]
    subset['change50'] = np.where(subset.perc_change_yr_50>change_thresh,1,0)
    means = subset[['depletion_scenario','metric','change50']].groupby(['metric','depletion_scenario']).sum()        
    means = means.reset_index()
    for metric in metric_list:
        ax = axs#[i]
        ax.boxplot(means[means.metric==metric].change50/len(metric_change.site.drop_duplicates()),
                  positions=[j],
                  boxprops= dict(linewidth=2.0, color='black'),
               whiskerprops = dict(linewidth=2.0, color='black'),
               medianprops = dict(linewidth=2.0, color='firebrick'),
               capprops = dict(linewidth=2.0, color='black'))
        ax.set_ylabel('Fraction of sites for which metric changed\nbetween year 1 and 50')

        j+=1
    i+=1
ax.set_xlabel('Metric')
ax.set_xticklabels(metric_list,rotation = 90)
ax.set_title('Annual flow metrics')
```
